```python
#Drop Null values
tvshows_rating = tvshows.copy().dropna(subset=['Certificate', 'Runtime_of_Episodes', 'Star3', 'Star4'])
```

```python
#Building the Recommender System
m = tvshows_rating['No_of_Votes'].quantile(0.75)
m
```

```python
C = tvshows_rating['No_of_Votes'].mean()
C
```

```python
def weighted_rating(df, m, C):
    term = df['No_of_Votes'] / (m + df['No_of_Votes'])
    return df['IMDB_Rating'] * term + (1-term) * C
```

```python
tvshows_rating["weighted_rating"] = tvshows_rating.apply(weighted_rating, axis=1, args=(m, C))
tvshows_rating.head()
```

```python
#Top 10 Action and Crime TV Shows
filter_genre = (tvshows_rating['Genre'].str.contains("Action", case=False) & 
                tvshows_rating['Genre'].str.contains("Crime", case=False))
filtered = tvshows_rating[filter_genre].sort_values(by=['IMDB_Rating'], ascending=False)
filtered.head(10)
```
